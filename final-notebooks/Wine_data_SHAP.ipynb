```python
#Load libraries

import pandas as pd
import numpy as np
np.random.seed(0)
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
import shap
import lime
import lime.lime_tabular
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB
import xgboost as xgb
from sklearn.svm import SVC
#print('done')
```

```python
#Load wine dataset

df = pd.read_csv('../data/datasets_4458_8204_winequality-red.csv') # Load the data

# Create Classification version of target variable
df['goodquality'] = [1 if x >= 7 else 0 for x in df['quality']]
# Separate feature variables and target variable
X = df.drop(['quality','goodquality'], axis = 1)
Y = df['goodquality']

# Split the data into train and test data:
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)

#print(X)
#print(Y)
```

```python
#Build the random forest classifier
rf = RandomForestClassifier(max_depth=6, random_state=0, n_estimators=10)
rf.fit(X_train, Y_train)

yPredRf = rf.predict(X_test)
print(metrics.classification_report(Y_test, yPredRf))
```

```python
#Start KNN implementation. Do elbow method to define number of K

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
# Number of k from 1 to 26
k_range = range(1, 26)
k_scores = []
# Calculate cross validation score for every k number from 1 to 26
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    # It’s 10 fold cross validation with ‘accuracy’ scoring 
    scores = cross_val_score(knn, X, Y, cv=10, scoring='accuracy') 
    k_scores.append(scores.mean())
    
%matplotlib inline
# Plot accuracy for every k number between 1 and 26
plt.plot(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-validated accuracy')
```

```python
# Train the model and use k=16
knn = KNeighborsClassifier(n_neighbors=16)
knn.fit(X_train, Y_train)
yPredKNN = knn.predict(X_test)

# classification report for test set
print(metrics.classification_report(Y_test, yPredKNN))
```

```python
#Train GNB model

gnb=GaussianNB()
gnb.fit(X_train,Y_train)

yPredGNB = gnb.predict(X_test)

# classification report for test set
print(metrics.classification_report(Y_test, yPredGNB))
```

```python
#Train XGB model

xgbModel = xgb.XGBClassifier(random_state=1)
xgbModel.fit(X_train, Y_train)
yPredXGB = xgbModel.predict(X_test)
print(metrics.classification_report(Y_test, yPredXGB))
```

```python
#Train SVM model with radial kernel

svclassifier = SVC(kernel='rbf')
svclassifier.fit(X_train, Y_train)

yPredSVM = svclassifier.predict(X_test)
print(metrics.classification_report(Y_test, yPredSVM))
```

```python
#Start SHAP implementation

#Use Kmeans to get summary dataset which reduces compute time for the kernel explainer
X_test_summary = shap.kmeans(X_test, 10)

#Create explainer and get SHAP values for RF
rf_explainer = shap.TreeExplainer(rf, feature_perturbation= "tree_path_dependent")
rf_shap_values = rf_explainer.shap_values(X_test)
print("rf")

#Create explainer and get SHAP values for KNN
knn_explainer = shap.KernelExplainer(knn.predict, X_test_summary)
knn_shap_values = knn_explainer.shap_values(X_test, nsamples = 100,l1_reg="num_features(11)")
print("knn")

#Create explainer and get SHAP values for GNB
gnb_explainer = shap.KernelExplainer(gnb.predict, X_test_summary)
gnb_shap_values = gnb_explainer.shap_values(X_test, nsamples = 100,l1_reg="num_features(11)")
print("gnb")

```

```python
#Update XGB model object in order to make it work with SHAP
myxgbbooster = xgbModel.get_booster()    
model_bytearray = myxgbbooster.save_raw()[4:]
def myfun(self=None):
    return model_bytearray
myxgbbooster.save_raw = myfun

#Create explainer and get SHAP values for XGB
xgb_explainer = shap.TreeExplainer(myxgbbooster, feature_perturbation= "tree_path_dependent")
xgb_shap_values = xgb_explainer.shap_values(X_test)
print("xgb")

#Create explainer and get SHAP values for SVM
svm_explainer = shap.KernelExplainer(svclassifier.predict, X_test_summary)
svm_shap_values = svm_explainer.shap_values(X_test, nsamples = 100,l1_reg="num_features(11)")
print("svm")
```

```python
#Plot all the summary plots for SHAP for all models

shap.summary_plot(rf_shap_values[1], X_test)
shap.summary_plot(knn_shap_values, X_test)
shap.summary_plot(gnb_shap_values, X_test)
shap.summary_plot(svm_shap_values, X_test)
shap.summary_plot(xgb_shap_values, X_test)
```

```python
#Force plot for specified index for all models

shap.initjs()
#Change the number below to observe local explanation for another point
index = 319
print("RF ", "Predicted: ", yPredRf[index], " ", "Actual: ", Y_test.values[index])
display(shap.force_plot(rf_explainer.expected_value[1], rf_shap_values[1][index,:], X_test.iloc[index,:]))
print("KNN ", "Predicted: ", yPredKNN[index], " ", "Actual: ", Y_test.values[index])
display(shap.force_plot(knn_explainer.expected_value, knn_shap_values[index,:], X_test.iloc[index,:], link='logit'))
print("GNB ", "Predicted: ", yPredGNB[index], " ", "Actual: ", Y_test.values[index])
display(shap.force_plot(gnb_explainer.expected_value, gnb_shap_values[index,:], X_test.iloc[index,:], link='logit'))
print("SVM ", "Predicted: ", yPredSVM[index], " ", "Actual: ", Y_test.values[index])
display(shap.force_plot(svm_explainer.expected_value, svm_shap_values[index,:], X_test.iloc[index,:], link='logit'))
print("XGB ", "Predicted: ", yPredXGB[index], " ", "Actual: ", Y_test.values[index])
display(shap.force_plot(xgb_explainer.expected_value, xgb_shap_values[index,:], X_test.iloc[index,:]))
#display(shap.waterfall_plot(knn_explainer.expected_value, knn_shap_values[291], feature_names=X.columns.values, max_display=7))
```
