```python
'''
# Install a pip package in the current Jupyter kernel
import sys
!{sys.executable} -m pip install shap
!{sys.executable} -m pip install lime
'''

import pandas as pd
import numpy as np
np.random.seed(0)
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.ensemble import RandomForestRegressor
import shap
import lime
import lime.lime_tabular

#print('done')
```

```python
df = pd.read_csv('D:/SDU/thesisCodes/imageClassification/data/datasets_4458_8204_winequality-red.csv') # Load the data
# The target variable is 'quality'.
Y = df['quality']
X =  df[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar','chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density','pH', 'sulphates', 'alcohol']]
# Split the data into train and test data:
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)

#print(X)
#print(Y)
```

```python

#Build an RF and then look at variable importance plot extracted from the default RF variable importance method
rf = RandomForestRegressor(max_depth=6, random_state=0, n_estimators=10)
rf.fit(X_train, Y_train)  
print(rf.feature_importances_)
importances = rf.feature_importances_
indices = np.argsort(importances)
features = X_train.columns
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()

```

```python
#Now try to build same plot with shap to see if there is any difference
shap_values = shap.TreeExplainer(rf).shap_values(X_test)

shap.summary_plot(shap_values, X_test, plot_type="bar")
```

```python
#A nicer and more intuitive plot
shap.initjs()
shap.summary_plot(shap_values, X_test)
```

```python
#Dependence plot example
shap.dependence_plot('alcohol', shap_values, X_test)
shap.dependence_plot('volatile acidity', shap_values, X_test)
```

```python
#Force plot for explaining individual instance
shap.initjs()
rf_explainer = shap.TreeExplainer(rf)
shap.force_plot(rf_explainer.expected_value, shap_values[19,:], X_test.iloc[19,:])
```

```python
#Collective force plot
shap.force_plot(rf_explainer.expected_value, shap_values, X_test)
```

```python
#Waterfall plot
shap.waterfall_plot(rf_explainer.expected_value[0], shap_values[19], feature_names=X.columns.values, max_display=7)
```

```python
X_featurenames = X.columns
explainer = lime.lime_tabular.LimeTabularExplainer(np.array(X_train),
                    feature_names=X_featurenames, 
                    class_names=['quality'], 
                    # categorical_features=, 
                    # There is no categorical features in this example, otherwise specify them.                               
                    verbose=True, mode='regression')
```

```python
#exp.show_in_notebook(show_table=True, show_all=True)

print(X_test.values[21])
print(X_test.iloc[21])

exp = explainer.explain_instance(X_test.values[21], rf.predict, num_features=11)
#plt.show(exp.as_pyplot_figure())
pd.DataFrame(exp.as_list())
```

```python
l=[]
densityList = []
fixedAcidityList = []
alcoholList = []
citricAcidList = []
residualSugrList = []
chloridesList = []
freeSulfurList = []
totalSulfurList = []
phList = []
volatileAcidityList = []
sulphatesList = []

limeDf = pd.DataFrame(columns=X_test.columns)
#limeDf

for n in range(0,320):
     exp = explainer.explain_instance(X_test.values[n], rf.predict, num_features=11)
     a=exp.as_list()
     l.append(a)

#print(l[1])
#pd.DataFrame(l)

for row in l:
    for elem in row:
        if 'density' in elem[0]:
            densityList.append(elem[1])
        elif 'alcohol' in elem[0]:
            alcoholList.append(elem[1])
        elif 'sulphates' in elem[0]:
            sulphatesList.append(elem[1])
        elif 'pH' in elem[0]:
            phList.append(elem[1])
        elif 'total' in elem[0]:
            totalSulfurList.append(elem[1])
        elif 'free' in elem[0]:
            freeSulfurList.append(elem[1])
        elif 'citric' in elem[0]:
            citricAcidList.append(elem[1])
        elif 'residual' in elem[0]:
            residualSugrList.append(elem[1])
        elif 'chlorides' in elem[0]:
            chloridesList.append(elem[1])
        elif 'fixed' in elem[0]:
            fixedAcidityList.append(elem[1])
        elif 'volatile' in elem[0]:
            volatileAcidityList.append(elem[1])

#print(dl)
limeDf['density'] = densityList
limeDf['volatile acidity'] = volatileAcidityList
limeDf['fixed acidity'] = fixedAcidityList
limeDf['citric acid'] = citricAcidList
limeDf['residual sugar'] = residualSugrList
limeDf['chlorides'] = chloridesList
limeDf['free sulfur dioxide'] = freeSulfurList
limeDf['total sulfur dioxide'] = totalSulfurList
limeDf['pH'] = phList
limeDf['sulphates'] = sulphatesList
limeDf['alcohol'] = alcoholList

limeDf
```
