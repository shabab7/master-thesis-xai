```python
#Load libraries

import pandas as pd
import numpy as np
np.random.seed(0)
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
import shap
import lime
import lime.lime_tabular
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB
import xgboost as xgb
from sklearn.svm import SVC
#print('done')
```

```python
#Load breast cancer dataset and do basic preprocessing
df = pd.read_csv('../data/datasets_180_408_data.csv') #

def impute_radius(cols):
    radius = cols[0]
    diagnosis = cols[1]
    
    # if value in radius column is null
    if pd.isnull(radius):
        
        # if woman is diagnosed with breast cancer
        if diagnosis == 1:
            return 17
        # if woman was not diagnosed with breast cancer
        else:
            return 12
    # when value in radius column is not null
    else:
        # return that same value
        return radius
    
df['radius_mean'] = df[['radius_mean', 'diagnosis']].apply(impute_radius, axis = 1)

df.drop(['id'], axis = 1, inplace = True)
df['diagnosis'] = [1 if x == 'M' else 0 for x in df['diagnosis']]
```

```python
#Further cleanup 
df.drop(['Unnamed: 32'], axis = 1, inplace = True)
df.shape
```

```python
#Define X and Y and also create train and test set

X = df.drop('diagnosis', axis = 1)
Y = df['diagnosis']

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.25, random_state = 42)
```

```python
#Build the random forest classifier and print metrics

rf = RandomForestClassifier(max_depth=6, random_state=0, n_estimators=10)
rf.fit(X_train, Y_train)

yPredRf = rf.predict(X_test)
print(metrics.classification_report(Y_test, yPredRf))
```

```python
#Start building KNN. First use elbow method to determine K value

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
# Number of k from 1 to 26
k_range = range(1, 26)
k_scores = []
# Calculate cross validation score for every k number from 1 to 26
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    # It’s 10 fold cross validation with ‘accuracy’ scoring 
    scores = cross_val_score(knn, X, Y, cv=10, scoring='accuracy') 
    k_scores.append(scores.mean())
    
%matplotlib inline
# Plot accuracy for every k number between 1 and 26
plt.plot(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-validated accuracy')
```

```python
#Train the KNN model and use K=14 based on previous graph

knn = KNeighborsClassifier(n_neighbors=14)
knn.fit(X_train, Y_train)
yPredKNN = knn.predict(X_test)

#classification report for test set
print(metrics.classification_report(Y_test, yPredKNN))
```

```python
#Train the GNB model

gnb=GaussianNB()
gnb.fit(X_train,Y_train)

yPredGNB = gnb.predict(X_test)

#classification report for test set
print(metrics.classification_report(Y_test, yPredGNB))
```

```python
#Train the XGB model

xgbModel = xgb.XGBClassifier(random_state=1)
xgbModel.fit(X_train, Y_train)
yPredXGB = xgbModel.predict(X_test)
print(metrics.classification_report(Y_test, yPredXGB))
```

```python
#Train the SVM model with radial kernel

svclassifier = SVC(kernel='rbf')
svclassifier.fit(X_train, Y_train)

yPredSVM = svclassifier.predict(X_test)
print(metrics.classification_report(Y_test, yPredSVM))
```

```python
#Train the SVM model with linear kernel

svclassifier = SVC(kernel='linear')
svclassifier.fit(X_train, Y_train)

yPredSVM = svclassifier.predict(X_test)
print(metrics.classification_report(Y_test, yPredSVM))
```

```python
#Start SHAP implementation

#Use Kmeans to get summary dataset which reduces compute time for the kernel explainer
X_test_summary = shap.kmeans(X_test, 10)

#Create explainer and get SHAP values for RF
rf_explainer = shap.TreeExplainer(rf, feature_perturbation= "tree_path_dependent")
rf_shap_values = rf_explainer.shap_values(X_test)
print("rf")

#Create explainer and get SHAP values for KNN
knn_explainer = shap.KernelExplainer(knn.predict, X_test_summary, l1_reg="num_features(30)")
knn_shap_values = knn_explainer.shap_values(X_test)
print("knn")

#Create explainer and get SHAP values for GNB
gnb_explainer = shap.KernelExplainer(gnb.predict, X_test_summary, l1_reg="num_features(30)")
gnb_shap_values = gnb_explainer.shap_values(X_test)
print("gnb")

```

```python
#Update XGB model object in order to make it work with SHAP
mybooster = xgbModel.get_booster()    
model_bytearray = mybooster.save_raw()[4:]
def myfun(self=None):
    return model_bytearray
mybooster.save_raw = myfun

#Create explainer and get SHAP values for XGB
xgb_explainer = shap.TreeExplainer(mybooster, feature_perturbation= "tree_path_dependent")
xgb_shap_values = xgb_explainer.shap_values(X_test)
print("xgb")

#Create explainer and get SHAP values for SVM
svm_explainer = shap.KernelExplainer(svclassifier.predict, X_test_summary, l1_reg="num_features(30)")
svm_shap_values = svm_explainer.shap_values(X_test)
print("svm")
```

```python
#Plot all the summary plots for SHAP for all models

shap.summary_plot(rf_shap_values[1], X_test)
shap.summary_plot(knn_shap_values, X_test)
shap.summary_plot(gnb_shap_values, X_test)
shap.summary_plot(svm_shap_values, X_test)
shap.summary_plot(xgb_shap_values, X_test)
```

```python
#Force plot for specified index for all models
#Also print the actual and predicted value for the specific row

shap.initjs()
#Change the number below to observe local explanation for another point
index = 8
print("RF ", "Predicted: ", yPredRf[index], " ", "Actual: ", Y_test.values[index])
display(shap.force_plot(rf_explainer.expected_value[1], rf_shap_values[1][index,:], X_test.iloc[index,:]))
print("KNN ", "Predicted: ", yPredKNN[index], " ", "Actual: ", Y_test.values[index])
display(shap.force_plot(knn_explainer.expected_value, knn_shap_values[index,:], X_test.iloc[index,:]))
print("GNB ", "Predicted: ", yPredGNB[index], " ", "Actual: ", Y_test.values[index])
display(shap.force_plot(gnb_explainer.expected_value, gnb_shap_values[index,:], X_test.iloc[index,:]))
print("SVM", "Predicted: ", yPredSVM[index], " ", "Actual: ", Y_test.values[index])
display(shap.force_plot(svm_explainer.expected_value, svm_shap_values[index,:], X_test.iloc[index,:]))
print("XGB", "Predicted: ", yPredXGB[index], " ", "Actual: ", Y_test.values[index])
display(shap.force_plot(xgb_explainer.expected_value, xgb_shap_values[index,:], X_test.iloc[index,:]))
```
