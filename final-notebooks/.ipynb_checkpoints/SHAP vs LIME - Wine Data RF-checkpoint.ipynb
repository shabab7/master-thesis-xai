```python
#Load libraries

import pandas as pd
import numpy as np
np.random.seed(0)
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.ensemble import RandomForestRegressor
import shap
import lime
import lime.lime_tabular

#print('done')
```

```python
#Load wine dataset

df = pd.read_csv('D:/SDU/thesisCodes/data/datasets_4458_8204_winequality-red.csv') # Load the data
# The target variable is 'quality'.
Y = df['quality']
X =  df[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar','chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density','pH', 'sulphates', 'alcohol']]
# Split the data into train and test data:
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)

#print(X)
#print(Y)
```

```python
#Build the random forest classifier
rf = RandomForestRegressor(max_depth=6, random_state=0, n_estimators=10)
rf.fit(X_train, Y_train)
```

```python
#Generate SHAP values

shap_values = shap.TreeExplainer(rf).shap_values(X_test)
```

```python
#Generate LIME values 1

X_featurenames = X.columns
explainer = lime.lime_tabular.LimeTabularExplainer(np.array(X_train),
                    feature_names=X_featurenames, 
                    class_names=['quality'], 
                    # categorical_features=, 
                    # There is no categorical features in this example, otherwise specify them.                               
                    verbose=True, mode='regression')
```

```python
#Generate LIME values 2

l=[]
densityList = []
fixedAcidityList = []
alcoholList = []
citricAcidList = []
residualSugrList = []
chloridesList = []
freeSulfurList = []
totalSulfurList = []
phList = []
volatileAcidityList = []
sulphatesList = []

limeDf = pd.DataFrame(columns=X_test.columns)
#limeDf

for n in range(0,320):
     exp = explainer.explain_instance(X_test.values[n], rf.predict, num_features=11)
     a=exp.as_list()
     l.append(a)

#print(l[1])
#pd.DataFrame(l)

for row in l:
    for elem in row:
        if 'density' in elem[0]:
            densityList.append(elem[1])
        elif 'alcohol' in elem[0]:
            alcoholList.append(elem[1])
        elif 'sulphates' in elem[0]:
            sulphatesList.append(elem[1])
        elif 'pH' in elem[0]:
            phList.append(elem[1])
        elif 'total' in elem[0]:
            totalSulfurList.append(elem[1])
        elif 'free' in elem[0]:
            freeSulfurList.append(elem[1])
        elif 'citric' in elem[0]:
            citricAcidList.append(elem[1])
        elif 'residual' in elem[0]:
            residualSugrList.append(elem[1])
        elif 'chlorides' in elem[0]:
            chloridesList.append(elem[1])
        elif 'fixed' in elem[0]:
            fixedAcidityList.append(elem[1])
        elif 'volatile' in elem[0]:
            volatileAcidityList.append(elem[1])

#print(dl)
limeDf['density'] = densityList
limeDf['volatile acidity'] = volatileAcidityList
limeDf['fixed acidity'] = fixedAcidityList
limeDf['citric acid'] = citricAcidList
limeDf['residual sugar'] = residualSugrList
limeDf['chlorides'] = chloridesList
limeDf['free sulfur dioxide'] = freeSulfurList
limeDf['total sulfur dioxide'] = totalSulfurList
limeDf['pH'] = phList
limeDf['sulphates'] = sulphatesList
limeDf['alcohol'] = alcoholList

limeDf
```

```python
#Fix LIME values and plot

absLimeDf = limeDf.abs()
meanLimeDf = pd.DataFrame(absLimeDf.mean(axis=0), columns=['Mean Lime Value'])
sortedLimeDf = meanLimeDf.sort_values('Mean Lime Value')
sortedLimeDf.plot.barh()
```

```python
#Plot SHAP values

shap.summary_plot(shap_values, X_test, plot_type="bar")
```
